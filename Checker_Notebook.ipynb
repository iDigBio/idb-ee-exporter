```python
(diff
 .select(col("uuid"))
 .distinct()
 .count()
)
```

```python
summary = (diff
          .select(year(col("datemodified")).alias("year"),
                  month(col("datemodified")).alias("month"),
                  dayofmonth(col("datemodified")).alias("day"),
                  col("op"))
 .groupBy(col("year"), col("month"), col("day"), col("op"))
 .count()
 .orderBy([col("year"), col("month"), col("day"), col("op")], ascending=[False, False, False])
)
```

```python
summary.show(1000, truncate=False)
```
